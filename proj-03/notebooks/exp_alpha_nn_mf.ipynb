```python
import os
import json

import pandas as pd
import matplotlib.pyplot as plt

from lib.util import *

plt.style.use("ggplot")
```

```python
df = pd.read_json("../data/preprocessed/preprocessed.json", orient="index")
with open("../data/preprocessed/param.json", "r") as f:
    params = json.load(f)
    print("params:", params)

user_n, item_n = params["user_n"], params["item_n"]
```

```python
for model_name in ["nn", "mf"]:
    for alpha in np.linspace(0.0, 1.0, 6):
        args = Args(exp_name="nn-mf-alpha", model=model_name, alpha=alpha)
        print(args)
        model, losses, results = train(
            df=df, user_n=user_n, item_n=item_n, args=args
        )
        print(results[-1])
        add_record(args, results[-1])
```

```python
records = []
for p in os.listdir("./log/nn-mf-alpha/"):
    with open(f"./log/nn-mf-alpha/{p}", "r") as f:
        d = json.load(f)
        records.append(pd.json_normalize(d))

df = pd.concat(records).sort_values("args.alpha").reset_index(drop=True)
```

```python
df.groupby(["args.model", "args.lmda"]).max("evaluations.Uplift@30")
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

for label, model_df in df.groupby(["args.model", "args.lmda"]):
    model_df.plot("args.alpha", "evaluations.Uplift@30", ax=ax, label=label)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

for label, model_df in df.groupby(["args.model", "args.lmda"]):
    model_df.plot("args.alpha", "evaluations.Accuracy@30", ax=ax, label=label)
```
